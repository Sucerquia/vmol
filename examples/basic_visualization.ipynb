In this notebook I show an overview of vmol, a wrap of vpython to create molecules visualizations. The idea is that you as a user, run and understand each one of the cells bellow.

```python
from vmol.view import VMolecule
from vmol.tests.variables4tests import GPAmol # xyz file
from ase.io import read

atoms = read(GPAmol)  # create ase.Atoms
vm = VMolecule(atoms)  #create the VMol master object
```

```python
vm.show_axis(); # show xyz axis
```

```python
vm.setting_canvas(background=[1, 1, 1]); # change scenes properties
```

### Atoms

```python
# Look at the oxigen atom in the left (the red one)
vm.hide_atom(5)
```

```python
# here it comes again
vm.show_atom(5)
```

```python
# execute this cell several times. keep looking at the oxigen atom
vm.update_atom(5, pos=vm.vatoms[5].pos + vm._asvector([0.25, 0, 0]))
```

```python
# move it back and leave it in pace
vm.update_atom(5, pos=vm.atoms[5].position)
```

**Note:** vmol has two similar objects that shouldn't be confused. one is atoms that corresponds to ase Atoms object. And the other is vatoms, which is a list of vpython spheres that represent the atoms in the scene.

### Bonds

```python
vm.add_bond(1, 2);  # create a bond between atoms 1 and 2, where atom 1 corresponds to vm.atoms[0]
```

```python
vm.add_bond(1, 3, radius=0.05, color=[0, 0,0 ]); # create a bond with certain properties
```

```python
# create several bonds at once. In this case, creates three bonds (1-4, 4-5, 5-6), which
# are set to be red, green and blue bonds with different thickness.
vm.add_bonds([1, 4, 5], [4, 5, 6], 
             radii=[0.05, 0.075, 0.1],
             colors=[[1, 0, 0], [0, 1, 0], [0, 0, 1]]);
```

```python
# In Vpython there is not a method to delete an object, to remove an object
# from the scene, it is necessary to hide it. In this example, one bond is hiden
vm.hide_bond(1, 2);
```

```python
vm.hide_all_bonds();
```

### Angles and dihedrals

```python
# let's add some bonds to make visualization easier
vm.add_bonds([1, 1, 4, 5], [3, 4, 5, 6]);
```

```python
# Angles are represented as surfaces in arches shape between the bonds that define them
# Here tha angle between atoms 3, 1 and 4, with 1 being the vertex
vm.add_angle(3, 1, 4, color=[1, 0, 0]);
```

```python
# As well as with bonds, you can hide angles
vm.hide_angle(3, 1, 4);
```

```python
# Dihedrals are pretty similar but the vertex is placed in the geometrical center of the
# torsion axis and the arch is orthogonal to that axes.
vm.add_dihedral(1, 4, 5, 6,  color=[0, 1, 0]);
```

```python
# Of Course, all what you create can be also hidden
vm.hide_dihedral(1, 4, 5, 6);
```

### Bonds, angles and dihedrals. All at once.

```python
# if you have a list with all the Degrees of freedon you want to show,
# and you don't want to to one by one creating them in the scene, you
# can use add_dof. For example:

dofs = [[1, 2], [1, 3], [1, 4], [4, 5], [5, 6], # distances
        [3, 1, 4], #dihedral
        [1, 4, 5, 6]]
for dof in dofs:
    vm.add_dof(dof)
```

```python
# This also works if you have 4-element-definition of your dof 
dofs = [[6, 7, 0, 0],
        [7, 8, 0, 0],
        [8, 9, 0, 0], # distances
        [4, 5, 6, 0], #dihedral
        [6, 7, 8, 9]]
for dof in dofs:
    vm.add_dof(dof)
```
